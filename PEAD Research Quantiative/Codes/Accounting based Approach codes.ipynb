```python
import pandas as pd

# Load 3 Excel files
link_table = pd.read_excel("3 Compustat CRSP Link.xlsx")
accounting = pd.read_excel("4 Accounting Data.xlsx")
returns = pd.read_excel("5 Stock returns.xlsx")
```

```python
#clean link table
link_clean = link_table[
    (link_table['Primary Link Marker'] == 'P') &
    (link_table['Link Type Codd'].isin(['LC', 'LU']))
][['gvkey', 'Historical CRSP PERMNO Link to COMPUSTAT Record']]

link_clean = link_clean.rename(columns={
    'Global Company Key': 'gvkey',
    'Historical CRSP PERMNO Link to COMPUSTAT Record': 'permno'
})
```

```python
# Merge accounting data with permno via gvkey

accounting_linked = pd.merge(accounting, link_clean, how='inner', on='gvkey')
```

```python
accounting_linked = accounting_linked.sort_values(by=['gvkey', 'datadate'])
# Raw EPS Surprise
accounting_linked['eps_lag'] = accounting_linked.groupby('gvkey')['eps'].shift(1)
accounting_linked['surprise_raw'] = accounting_linked['eps'] - accounting_linked['eps_lag']
# Rolling std over last 4 quarters for SUE
accounting_linked['eps_std_4q'] = accounting_linked.groupby('gvkey')['eps'].rolling(4).std().reset_index(0, drop=True)
accounting_linked['sue'] = accounting_linked['surprise_raw'] / accounting_linked['eps_std_4q']
```

```python
# Convert dates
returns['names date'] = pd.to_datetime(returns['Names Date'])
accounting_linked['rdq'] = pd.to_datetime(accounting_linked['rdq'])
```

```python
import pandas as pd
import numpy as np

# Load files
link_table = pd.read_excel("3 Compustat CRSP Link.xlsx")
accounting = pd.read_excel("4 Accounting Data.xlsx")
returns = pd.read_excel("5 Stock returns.xlsx")

# Clean link table
link_clean = link_table[
    (link_table['Primary Link Marker'] == 'P') &
    (link_table['Link Type Codd'].isin(['LC', 'LU']))
][['gvkey', 'Historical CRSP PERMNO Link to COMPUSTAT Record']]
link_clean = link_clean.rename(columns={
    'Global Company Key': 'gvkey',
    'Historical CRSP PERMNO Link to COMPUSTAT Record': 'permno'
})
```

```python
# Merge and sort
accounting_linked = pd.merge(accounting, link_clean, how='inner', on='gvkey')
accounting_linked['rdq'] = pd.to_datetime(accounting_linked['rdq'])
accounting_linked = accounting_linked.sort_values(by=['gvkey', 'rdq'])
```

```python
# EPS lag
accounting_linked['eps_lag'] = accounting_linked.groupby('gvkey')['eps'].shift(1)
accounting_linked['surprise_raw'] = accounting_linked['eps'] - accounting_linked['eps_lag']
```

```python
# Rolling std for 4Q and 8Q
accounting_linked['eps_std_4q'] = (
    accounting_linked.groupby('gvkey')['eps']
    .rolling(window=4, min_periods=2).std()
    .reset_index(level=0, drop=True)
)
accounting_linked['eps_std_8q'] = (
    accounting_linked.groupby('gvkey')['eps']
    .rolling(window=8, min_periods=4).std()
    .reset_index(level=0, drop=True)
)
```

```python
# Avoid divide by zero
eps_std_4q = accounting_linked['eps_std_4q'].replace(0, np.nan)
eps_std_8q = accounting_linked['eps_std_8q'].replace(0, np.nan)

# Compute SUEs
accounting_linked['sue_4q'] = accounting_linked['surprise_raw'] / (eps_std_4q + 1e-6)
accounting_linked['sue_8q'] = accounting_linked['surprise_raw'] / (eps_std_8q + 1e-6)
```

```python
# Load & prep returns
returns.columns = returns.columns.str.lower().str.strip()
returns = returns.rename(columns={'names date': 'event_date'})
returns['event_date'] = pd.to_datetime(returns['event_date'])
returns = returns.loc[:, ~returns.columns.duplicated()]
```

```python
# Create event structure: ±60 days around rdq
event_rows = []
for _, row in accounting_linked.iterrows():
    for day in range(-60, 61):
        event_date = row['rdq'] + pd.Timedelta(days=day)
        event_rows.append({
            'gvkey': row['gvkey'],
            'permno': row['permno'],
            'rdq': row['rdq'],
            'event_day': day,
            'event_date': event_date
        })
event_df = pd.DataFrame(event_rows)
```

```python
# Merge event structure with returns
merged = pd.merge(event_df, returns[['permno', 'event_date', 'returns']],
on=['permno', 'event_date'], how='left')

# Merge in accounting SUEs
final_panel = pd.merge(
    merged,
    accounting_linked[['gvkey', 'rdq', 'eps', 'eps_lag', 'surprise_raw',
'eps_std_4q', 'sue_4q', 'eps_std_8q', 'sue_8q']],
    on=['gvkey', 'rdq'],
    how='left'
)

# Filter extreme returns
final_panel = final_panel.dropna(subset=['returns'])
final_panel = final_panel[(final_panel['returns'] > -5) & (final_panel['returns'] < 5)]
# Export
final_panel.to_excel("sue_panel_new.xlsx", index=False)
import pandas as pd

# Load the SUE panel created earlier
sue_panel = pd.read_excel("sue_panel_new.xlsx")

# Make sure rdq is datetime and create a 'quarter' column for grouping
sue_panel['rdq'] = pd.to_datetime(sue_panel['rdq'])
sue_panel['quarter'] = sue_panel['rdq'].dt.to_period('Q')

# print unique quarters
print("Quarter coverage:", sue_panel['quarter'].nunique(), "quarters")
# Define a function to compute deciles within each quarter
def assign_deciles(df, column, decile_col):
    df[decile_col] = (
        df.groupby('quarter')[column]
        .transform(lambda x: pd.qcut(x.rank(method='first'), 10, labels=False, duplicates='drop') + 1)
    )
    return df

# Apply to both sue_4q and sue_8q
sue_panel = assign_deciles(sue_panel, 'sue_4q', 'decile_4q')
sue_panel = assign_deciles(sue_panel, 'sue_8q', 'decile_8q')

# Check that the deciles are assigned correctly
print("4Q Decile Distribution:\n", sue_panel['decile_4q'].value_counts().sort_index())
print("\n8Q Decile Distribution:\n", sue_panel['decile_8q'].value_counts().sort_index())
```

```python
# Required Libraries
import pandas as pd
import numpy as np

# Load Data
df = pd.read_excel("merged_panel_sue_ff3.xlsx")  # or whatever your filename is

# Clean & Prepare
df["rdq"] = pd.to_datetime(df["rdq"], errors="coerce")
df["event_date"] = pd.to_datetime(df["event_date"], errors="coerce")
df["returns"] = pd.to_numeric(df["returns"], errors="coerce")
df["event_day"] = pd.to_numeric(df["event_day"], errors="coerce")

# Drop rows with missing key values
df = df.dropna(subset=["permno", "rdq", "event_day", "returns", "decile_8q"])

# Check if we have any post-60 day data
print("Event days > 60:", df[df["event_day"] > 60].shape[0])
print("Sample post-60 rows:")
print(df[df["event_day"] > 60][["permno", "rdq", "event_day", "returns"]].head())

# Assign future return buckets
df["ret_3m"] = np.where(df["event_day"].between(61, 120), df["returns"], 0)
df["ret_6m"] = np.where(df["event_day"].between(121, 180), df["returns"], 0)
df["ret_12m"] = np.where(df["event_day"].between(181, 252), df["returns"], 0)

# Collapse into cumulative horizon returns
df_horizon = df.groupby(["permno", "rdq", "decile_8q"])[["ret_3m", "ret_6m", "ret_12m"]].sum().reset_index()

# Mean returns by decile
df_summary = df_horizon.groupby("decile_8q")[["ret_3m", "ret_6m", "ret_12m"]].mean().reset_index().sort_values("decile_8q")

# Show results
print("Mean Post-Event Returns by SUE Decile")
print(df_summary)

import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# ===Load your data ===
df = pd.read_excel("merged_panel_sue_ff3.xlsx")

# === Filter event window [+1, +60] for CAR ===
event_window = df[(df['event_day'] >= 1) & (df['event_day'] <= 60)].copy()

# === Compute CAR per firm-quarter ===
# Calculate excess return = returns - RF
event_window['excess_return'] = event_window['returns'] - event_window['RF']

# Aggregate CAR = sum of excess returns for each firm-quarter
car_df = (
    event_window
    .groupby(['permno', 'quarter'])  # firm-quarter level
    .agg({
        'excess_return': 'sum',
        'sue_8q_x': 'first'  # same across the 60-day window
    })
    .rename(columns={'excess_return': 'CAR', 'sue_8q_x': 'SUE'})
    .dropna()
    .reset_index()
)

# ===Run cross-sectional regression in each quarter ===
quarter_betas = []

for q in car_df['quarter'].unique():
    sub = car_df[car_df['quarter'] == q]
    if len(sub) < 10:  # skip small samples
        continue

    X = sm.add_constant(sub['SUE'])
    y = sub['CAR']
    model = sm.OLS(y, X).fit()
    quarter_betas.append(model.params)

# Convert results into DataFrame
beta_df = pd.DataFrame(quarter_betas)
beta_df.columns = ['alpha', 'beta_SUE']

# ===Summary of Fama-MacBeth results ===
mean_beta = beta_df['beta_SUE'].mean()
std_beta = beta_df['beta_SUE'].std()
t_stat = mean_beta / (std_beta / (len(beta_df) ** 0.5))

print("=== Fama-MacBeth Regression Results ===")
print(f"Mean Beta: {mean_beta:.4f}")
print(f"Std. Dev of Betas: {std_beta:.4f}")
print(f"T-statistic: {t_stat:.3f}")
print(f"Number of Quarters: {len(beta_df)}")

# === Plot beta estimates over time ===
plt.figure(figsize=(10, 5))
plt.plot(beta_df['beta_SUE'], marker='o')
plt.title(r"$\it{Figure\ 18:\ SUE\ Coefficient\ Estimates\ (Quarterly\ Cross\ Section)}$",
          x=0.5,    # Center the title
          pad=20,   # Add 20pt space above plot
          fontsize=12)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel("Quarter Index")
plt.ylabel("Beta on SUE")
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# === STEP 1: LIBRARIES ===
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# === STEP 2: LOAD DATA ===
# Adjust the filename/path if needed
df = pd.read_excel("merged_panel_sue_ff3.xlsx")

# === STEP 3: FILTER FOR EVENT WINDOW [+1, +60] ===
df_8q = df[(df['event_day'] >= 1) & (df['event_day'] <= 60)]

# === STEP 4: CALCULATE FIRM-LEVEL CAR (Cumulative Abnormal Return) ===
car_8q = df_8q.groupby(['gvkey', 'decile_8q'])['returns'].sum().reset_index()

# === STEP 5: ISOLATE TOP (D10) AND BOTTOM (D1) DECILES ===
car_d10_8q = car_8q[car_8q['decile_8q'] == 10]['returns']
car_d1_8q  = car_8q[car_8q['decile_8q'] == 1]['returns']

# === STEP 6: ONE-SIDED T-TEST: H1 = D10 > D1 ===
t_stat_8q, p_val_8q = ttest_ind(car_d10_8q, car_d1_8q, alternative='greater')



# === STEP 8: PRINT RESULTS ===
print("=== Hypothesis Test: CAR_D10 > CAR_D1 (8Q SUE) ===")
print(f"Mean CAR - D10: {car_d10_8q.mean():.4f}")
print(f"Mean CAR - D1: {car_d1_8q.mean():.4f}")
print(f"T-statistic: {t_stat_8q:.4f}")
print(f"One-sided p-value: {p_val_8q:.4f}")
print(f"Sample sizes → D10: {len(car_d10_8q)}, D1: {len(car_d1_8q)}")
```

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Step 1: Load the dataset
df_4q = pd.read_excel('merged_panel_sue_ff3.xlsx')

# Step 2: Basic sanity check
print("Data Preview:")
print(df_4q[['permno', 'quarter', 'returns', 'sue_4q']].head())

# Step 3: Drop missing values
df_4q_clean = df_4q[['permno', 'quarter', 'returns', 'sue_4q']].dropna()
print("\nData shape after dropping NA:", df_4q_clean.shape)

# Step 4: Rename for convenience
df_4q_clean = df_4q_clean.rename(columns={'returns': 'CAR', 'sue_4q': 'SUE'})

# Step 5: Convert firm & time to categorical for FE
df_4q_clean['firm_fe'] = df_4q_clean['permno'].astype('category')
df_4q_clean['time_fe'] = df_4q_clean['quarter'].astype('category')

# Step 6: Run Fixed Effects Model using Patsy-style formula
model_fe_4q = smf.ols(formula='CAR ~ SUE + C(firm_fe) + C(time_fe)', data=df_4q_clean).fit()

# Step 7: Output summary
print("\n=== Fixed Effects Regression: 4Q SUE ===")
print(model_fe_4q.summary())

# Optional: Check R-squared and coefficient
print("\nCoefficient on SUE:", round(model_fe_4q.params['SUE'], 4))
print("P-value on SUE:", round(model_fe_4q.pvalues['SUE'], 4))
print("R-squared:", round(model_fe_4q.rsquared, 4))
```
